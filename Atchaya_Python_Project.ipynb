```python
import matplotlib.pyplot as plt

genre_count = df['Genre'].value_counts()

plt.figure()
genre_count.plot(kind='bar',color='orange')
plt.title("Number of Movies by Genre")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
df['Rating'].plot(kind='hist', bins=10)
plt.title("Movie Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
avg_revenue = df.groupby('Genre')['Revenue_Millions'].mean()
avg_revenue.plot(kind='bar',color='green')
plt.title("Average Revenue by Genre")
plt.xlabel("Genre")
plt.ylabel("Revenue (Millions)")
plt.show()
```

```python
plt.figure()
movies_per_year = df['Year'].value_counts().sort_index()
movies_per_year.plot(kind='line', marker='o',color='yellow')
plt.title("Movies Released by Year")
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.show()
```

```python
plt.figure()
plt.scatter(df['Rating'], df['Revenue_Millions'])
plt.title("Rating vs Revenue")
plt.xlabel("Rating")
plt.ylabel("Revenue (Millions)")
plt.show()
```

```python
plt.figure()
top_directors = df['Director'].value_counts().head(10)
top_directors.plot(kind='bar',color='red')
plt.title("Top 10 Directors by Movies")
plt.xlabel("Director")
plt.ylabel("Number of Movies")
plt.show()
```
